```python
from pathlib import Path
import csv

PROJECT_ROOT = Path.cwd().parents[1] if Path.cwd().name == 'notebooks' else Path.cwd()
GOLD_SYMBOLS = ['GLD', 'IAU', 'GDX', 'GDXJ', 'SLV', 'NEM', 'AEM', 'GOLD', 'FNV', 'WPM', 'KGC', 'AGI', 'PAAS', 'FCX']
COMPARISON_SECTORS = ['Cloud / SaaS', 'Semiconductors', 'Healthcare', 'Financials']

CACHE_ROOTS = [
    PROJECT_ROOT / 'data' / 'historical' / 'yahoo_2022_2026' / 'ohlcv',
    PROJECT_ROOT / 'data' / 'historical' / 'ohlcv',
    PROJECT_ROOT / 'data' / 'historical' / 'massive_2022_2026' / 'ohlcv',
]
```

```python
def load_bars(symbol):
    candidates = []
    for root in CACHE_ROOTS:
        for path in root.glob(f'{symbol}_*.csv'):
            with path.open(newline='', encoding='utf-8') as file:
                rows = list(csv.DictReader(file))
            if rows:
                candidates.append((path, rows))
    if not candidates:
        return None, []
    return max(candidates, key=lambda item: len(item[1]))

coverage = []
for symbol in GOLD_SYMBOLS:
    path, rows = load_bars(symbol)
    coverage.append({
        'symbol': symbol,
        'bars': len(rows),
        'start': rows[0]['date'] if rows else None,
        'end': rows[-1]['date'] if rows else None,
        'source': path.parent.parent.name if path else None,
    })

coverage
```
